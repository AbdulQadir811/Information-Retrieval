##Task 2
Write a function that reurns a dataframe with two columns having stop words in one col and their count in the other.

```python
import re
import pandas as pd
def findWholeWord(w):
    return re.compile(r'\b({0})\b'.format(w), flags=re.IGNORECASE).search


def CreateDF(filename):
  words_list=[]
  count_list=[] 
  mydict={}
     
  f  = open(filename,'rb')
  data = f.read()
  f.close()  
  
  swf  = open('stopwords.txt','rb')
  sw_data = swf.read()  
  swf.close()  
  
  stop_words_list = (str(sw_data).split("\\n"))
  stop_words_list[0]=stop_words_list[0][2:]
  stop_words_list[len(stop_words_list)-1]=stop_words_list[0][:len(stop_words_list)]
  
  lines = (str(data).split("\\r\\n"))
  lines[0]=lines[0][2:]
  
  for word in stop_words_list:
    mydict[word] = 0


  for line in lines:
        line = line.lower()
        line= line.replace("\\'","'")
        line= line.replace(".","")
        line= line.replace(",","")
        for word in stop_words_list:
            try:
                if findWholeWord(word)(line) is not None:
                    mydict[word]=(mydict[word]) + 1
            except Exception as e:
                print("[!] ",e )
  
  word_list = mydict.keys()
  count_list = mydict.values()  
  df = pd.DataFrame(list(zip(word_list,count_list)),columns=["word","count"])
  df.to_csv("result.csv",index=False)   
  return df


CreateDF('Text_File.txt')
```

```python
fileName="Text_File.txt"
df = CreateDF(fileName)
df
```

##Task 4
Write a function that displays all words having length greater than 5.

```python
def GreaterThan5(fileName):
        
        f  = open(fileName,'rb')
        data = f.read()  
        f.close() 
        
        lines = (str(data).split("\\r\\n"))
        lines[0]=lines[0][2:]
        for line in lines:
            line= line.replace("\\'","'")
            line= line.replace(".","")
            line = line.replace(",","")
            greaterthan5= [i for i in line.split() if len(i.strip()) > 5]
            if len(greaterthan5) >1:
                print(greaterthan5)

                

GreaterThan5("Text_File.txt")
```

##Task 5
Write a function that returns the count of all words which end at y.

```python
def EndAty(fileName):
        yEnding = 0
        
        f  = open(fileName,'rb')
        data = f.read()
        f.close()  
        
        lines = (str(data).split("\\r\\n"))
        lines[0]=lines[0][2:]
        
        for line in lines:
            line= line.replace("\\'","'")
            line= line.replace(".","")
            line= line.replace(",","")
            greaterthan5= [i for i in line.split() if ord(i[-1].lower()) > 9 and ord(i[-1].lower()) < 122 and i[-1].lower() =="y"]
            if len(greaterthan5) > 0:
                yEnding= yEnding+len(greaterthan5)
        return yEnding
filename="Text_File.txt"
EndAty(filename)
```

##Task 7 
Write a function to replace all "the" with "thee".

```python

def ReplaceThe(fileName):
        f  = open(fileName,'rb')
        data = f.read()
        f.close()  
        lines = (str(data).split("\\r\\n"))
        lines[0]=lines[0][2:]
        lines.pop()[-1]
        updated_file  =  open("thee_text_file.txt",'w')
        for line in lines:
            line= line.replace("\\'","'")
            line= line.replace(",","")
            line= line.replace(".","")
            line =line.replace("\ \''","\'")
            line = line.split()
            
            len_3_words= [location for location,i in enumerate(line) if len(i)==3]
           
            if len(len_3_words) > 0:
                for i,index in enumerate(len_3_words):
                    if "the"==line[index].lower():
                        line[index] = line[index]+ "e"
            delimiter = ' '
            line  = delimiter.join(line)
            updated_file.write(str(line)+"\n")
            print(line)
           
        return None
    
filename="Text_File.txt"


ReplaceThe(filename)

```

##Task 8
Write a function ZICount() that prints the count of all occurences of Z and I (including small cases a and i too)
For example: Hi! I am Zenia. It was so good to see you! 
####Output: 
I or i : 4

Z or z : 1


```python
def ZICount(filename):
    f  = open(filename,'rb')
    data = f.read()
    f.close()  
    z_count  = 0
    i_count  = 0
    for letter in data:
        if letter == 90 or letter == 122:
            z_count = z_count + 1
        elif z_count == 73 or letter == 151:
            i_count= i_count + 1
    result ={"i_count" : i_count,"z_count":z_count}
    return result

ZICount('Text_File.txt')
```

##Task 10
10.	Write a function alter_Upper that prints the content in a way that the first line is in Upper case, second in lower case and so on.

Example: 
I am Zenia
I love painting
I live in Pakistan

Output:
i am zenia
I LOVE PAINTING
i live in pakistan


```python
def alter_Upper(filename):
    with open(filename,errors= "ignore") as fp:
        data = fp.readlines()
    for line in data :
        new_string = ""
        old_string = "" 
        i = 0
        while i < len(line):
            asci = ord(line[i])
            if asci > 64 and  asci < 91:
                new_string = new_string + chr(asci+32)       
            elif asci > 96 and  asci < 122:
                new_string = new_string + chr(asci-32) 

            else:
                new_string = new_string+line[i]
            old_string = old_string + chr(asci)
            i = i + 1
        print("new string ",new_string)
        print("old string ",old_string)
           
        
```
